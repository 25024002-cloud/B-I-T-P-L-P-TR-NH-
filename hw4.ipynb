W4A18 Nhập vào số nguyên dương a và b, in toàn bộ ước chung của a và b 

```python
a, b = map(int, input().split())
nho_nhat = min(a, b)
for i in range(1, nho_nhat + 1):
    if a % i == 0 and b % i == 0:
        print(i)
```

W4A21 Nhập vào số nguyên dương n, tính tổng các chữ số của n

```python
n = int(input())
tong = 0
while n > 0:
    chu_so = n % 10
    tong = tong + chu_so
    n //= 10
print(tong)
```

W4A22 Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ 


```python
n = int(input())
chan = 0
le = 0
while n > 0:
    chu_so = n % 10
    if chu_so % 2 == 0:
        chan += 1
    else:
        le += 1
    n //= 10
print(chan, le)
```

W4A25
Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1.
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập 


```python
n = int(input())
if n == -1:
    pass # Không có số nào được nhập
else:
    min_val = max_val = n
    while True:
        n = int(input())
        if n == -1: break
        if n < min_val: min_val = n
        if n > max_val: max_val = n
    print(max_val, min_val)
```

W4A26
[Fibonaci1]
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ
bằng tổng hai số trước đó
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt
quá A 


```python
a = int(input())
f1, f2 = 1, 1
while f1 + f2 <= a:
    f_tiep_theo = f1 + f2
    f1 = f2
    f2 = f_tiep_theo
print(f2)
```

W4A29
Cho một chuỗi chỉ gồm các ký tự chữ cái thường và chữ cái hoa, hãy đổi ký tự hoa thành 

```python
s = input()
res = ""
for char in s:
    if 'a' <= char <= 'z':
        res += chr(ord(char) - 32)
    elif 'A' <= char <= 'Z':
        res += chr(ord(char) + 32)
    else:
        res += char
print(res)
```

W4A31
Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in 

```python
s = input()
hoa = thuong = so = 0
for char in s:
    if char.isupper(): hoa += 1
    elif char.islower(): thuong += 1
    elif char.isdigit(): so += 1
print(hoa, thuong, so)
```
