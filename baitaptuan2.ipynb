W2A4:Viết chương trình python để tính điểm trung bình môn học của một học sinh trong lớp. Có tổng cộng
6 đầu điểm: a1, b1, c1, a2, b2, a3.
Công thức tính điểm trung bình là: TB=((a1+b1+c1)+(a2+b2)×2+a3×3)/10.Đầu ra: Đầu ra là giá trị điểm trung bình môn học của 6 môn học, lấy đến 1 chữ số phần thập phân.


```python
# Nhập 6 số trên 1 dòng
a1, b1, c1, a2, b2, a3 = map(float, input().split())
tb = ((a1 + b1 + c1) + (a2 + b2) * 2 + a3 * 3) / 10
print(f"tb = {tb:.1f}")
```
