Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng
giữ nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả.
Yêu cầu in ra đúng dạng list Python, ví dụ [1, 2, 3]


```python
lst = list(map(int, input().split()))
result = []
seen = set()
for i in lst:
    if i not in seen:
        result.append(i)
        seen.add(i)
print(result)

```

***W6A4 Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo
dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key
lặp lại, thêm vào list của key đó

```python
cac_cap = input().split()
anh_xa = {}
for cap in cac_cap:
    if ':' in cap:
        key, value = cap.split(':')
        if key in anh_xa:
            anh_xa[key].append(value)
        else:
            anh_xa[key] = [value]
print(anh_xa)
```

W6A5 Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ...,
'zeros': ...} (đếm số >0, <0, =0)


```python
lst = list(map(int, input().split()))
dictionary = {'positives':0,'negatives':0,'zeros':0}
for i in lst:
    if i>0:
        dictionary['positives'] += 1
    if i<0:
        dictionary['negatives'] += 1
    if i==0:
        dictionary['zeros'] += 1
print(dictionary)
    


```

W6A10 Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên
dương k. In ra list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là
một tuple, hai số trong tuple có thể giống nhau, thứ tự xuất hiện trong danh
sách của số thứ nhất nhỏ hơn thứ tự của số thứ hai. Các tuple trong list được
sắp xếp theo thứ tự tăng dần của phần tử đầu tiên trong tuple.

```python
lst = list(map(int, input().split()))
k = int(input())
kqua =[]
for i in len(lst):
    for j in range(i, len(lst)):
        if lst[i]+lst[j]==k:
            kqua.append((lst[i], lst[j]))
kqua.sort()
print(kqua)
```

W6A12 [Tìm phần tử xuất hiện nhiều nhất]
Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất trong
list. Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ
nhất trong số đó.


```python
lst = list(map(int, input().split()))
count={}
for so in lst:
    if so in count:
        count[so] += 1
    else:
        count[so] = 1
max = 0
for j in count.values():
    if j > max:
        max = j
ung_cu_vien = []
for k in count:
    if count[k] == max:
        ung_cu_vien.append(k)
ung_cu_vien.sort()
print(ung_cu_vien[0])
```

W6A13 [Đảo ngược key và value của dictionary]
Nhập vào một dictionary có value là duy nhất (không trùng nhau). Hãy tạo
một dictionary mới bằng cách đảo ngược key và value của dictionary ban
đầu. In ra dictionary mới.


```python
line1 = input().split()
dic1 = {}
for cap in line1:
    k, v = cap.split(':')
    dic1[k] = v
dic2 = {}
for k in dic1:
    key_cu = dic1[k]
    dic2[key_cu] = k
print(dic2)
```

W6A14 [Giao của hai danh sách]
