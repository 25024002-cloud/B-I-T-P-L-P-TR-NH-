W5A3 Viết hàm kiểm tra một số có phải là số nguyên tố hay không?

```python
def is_prime(n: int) -> bool:
    if n <= 1:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True
n = int(input())
print(is_prime(n))
```

W5A4 VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên. 


```python
def is_perfect(n: int) -> bool:
    if n <= 1:
        return False
    tong = 0
    for i in range(1, n):
        if n % i == 0:
            tong = tong + i
    if tong == n:
        return True
    else:
        return False
n = int(input())
print(is_perfect(n))
```

W5A6 Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!)

W5A8 :Khoảng cách Hamming giữa hai số nguyên là số lượng vị trí khác nhau giữa hai dãy bits tương ứng
của chúng. Ví dụ: Khoảng cách Hamming giữa 1 (1) và 4 (100) là 2.
Cho hai số nguyên x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này. 


```python
def hamming_distance(x: int, y: int) -> int:
    khoang_cach = 0
    
    # Chừng nào một trong hai số vẫn còn lớn hơn 0
    while x > 0 or y > 0:
        # Bước 1: Lấy bit cuối của x và y (số dư khi chia cho 2)
        bit_x = x % 2
        bit_y = y % 2
        
        # Bước 2: Nếu hai bit khác nhau thì tăng biến đếm
        if bit_x != bit_y:
            khoang_cach += 1
            
        # Bước 3: Chia nguyên cho 2 để xét cặp bit tiếp theo
        x = x // 2
        y = y // 2
        
    return khoang_cach

# --- Phần thực thi đầy đủ ---
x, y = map(int, input().split())
print(hamming_distance(x, y))
```

W5A15 Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông.

```python
def is_right_triangle(a, b, c):
    if a**2 + b**2 == c**2 or a**2 + c**2 == b**2 or c**2 + b**2 == a**2 :
        return True
    else:
        return False
a, b, c = map(float, input().split())
print(is_right_triangle(a, b, c))
```
