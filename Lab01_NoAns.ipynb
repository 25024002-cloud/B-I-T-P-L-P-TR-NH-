### Problem 4
#### Write a function that returns the count of positive integers k such that 1 <= k < n and k > (n - 1) / 2. (Example: n=9 -> 4).

```python
def dem_so_k(n):
    dem = 0
    nguong = (n-1)/2
    for k in range (1, n):
        if k > nguong:
            dem = dem + 1
    return dem
    
```

### Problem 8
#### Write a function that returns the character that is n positions after character x in the alphabet.
#### Preserve case and wrap around within A..Z or a..z.

```python
def shift_character(x, n):
    if 'a' <= x <= 'z':
        # Công thức cho chữ thường
        return chr((ord(x) - ord('a') + n) % 26 + ord('a'))
    
    if 'A' <= x <= 'Z':
        # Công thức cho chữ hoa
        return chr((ord(x) - ord('A') + n) % 26 + ord('A'))
    
    return x # Trả về chính nó nếu không phải chữ cái
```

### Problem 12
#### Write a function that returns a right isosceles triangle of height n using '*'.
#### Line i (1-indexed) has i '*', joined with newlines.
#### Example with n = 4
#### *
#### **
#### ***
#### ****

```python
def draw_triangle(n):
    ket_qua = ""
    for i in range(1, n + 1):
        # 1. Tạo ra một dòng có i dấu sao
        dong_sao = "*" * i
        
        # 2. Cộng dòng đó vào kết quả và thêm lệnh xuống dòng (\n)
        ket_qua = ket_qua + dong_sao + "\n"
        
    return ket_qua
```
