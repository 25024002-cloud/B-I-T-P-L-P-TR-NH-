W5A16: Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x


```python
def is_armstrong (x:int) -> bool :
    if x < 0:
        return False
    chuoi_x = str(x)
    so_chu_so = len(chuoi_x)
    tong = 0
    for chu_so in chuoi_x:
        so = int(chu_so)
        luy_thua = so**so_chu_so
        tong = tong + luy_thua
    if tong == x:
        return True
    else:
        return False
        
```

W5A19 VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên.


```python
def is_ki_dieu(x):
    if x<=0:
        return False
    while x >= 10:
        tong = 0
        while x>0:
            chu_so = x % 10
            tong = tong + chu_so
            x = x// 10
        x = tong
    return x == 1
```

W5A24 Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không.


```python
def are_collinear(x1, y1, x2, y2, x3, y3):
    if (x1==x2 and y1==y2) or (x1==x3 and y1==y3) or (x2==x3 and y2==y3):
        return False
     #Công thức: (y2-y1)/(x2-x1) == (y3-y1)*(x3-x1)
    if (y2 - y1) * (x3 - x1) == (y3 - y1) * (x2 - x1):
        return True
    else:
        return False
```

W5A25
Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay
không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3),
(0,3)], p = (2,1)

```python
def point_in_quadrilateral(quad, p):
    # quad là danh sách 4 đỉnh: [A, B, C, D]
    # p là điểm cần check: (px, py)
    
    def check_side(A, B, P):
        # Tính toán để xem P nằm bên trái hay bên phải đường thẳng AB
        return (B[0] - A[0]) * (P[1] - A[1]) - (B[1] - A[1]) * (P[0] - A[0])
    # Lấy tọa độ 4 đỉnh ra cho dễ nhìn
    A, B, C, D = quad
    # Kiểm tra P với từng cạnh: AB, BC, CD, DA
    d1 = check_side(A, B, p)
    d2 = check_side(B, C, p)
    d3 = check_side(C, D, p)
    d4 = check_side(D, A, p)

    # Nếu tất cả d1, d2, d3, d4 cùng >= 0 (hoặc cùng <= 0)
    # thì điểm p nằm trong hoặc trên biên của tứ giác.
    if (d1 >= 0 and d2 >= 0 and d3 >= 0 and d4 >= 0) or \
       (d1 <= 0 and d2 <= 0 and d3 <= 0 and d4 <= 0):
        return True
    else:
        return False
```

W5A28 Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).


```python
import math

def calculate_angle(x1, y1, x2, y2, x3, y3):
    # Tính góc của 2 cạnh BA và BC (lấy B làm gốc)
    goc1 = math.atan2(y1 - y2, x1 - x2)
    goc2 = math.atan2(y3 - y2, x3 - x2)
    
    # Tính độ lệch giữa 2 góc
    chenh_lech = math.degrees(goc2 - goc1)
    
    # Lấy giá trị dương và nhỏ hơn 180 độ
    ket_qua = abs(chenh_lech)
    if ket_qua > 180:
        ket_qua = 360 - ket_qua
        
    return ket_qua
```

Viết hàm smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính tổng tiền taxi theo quy tắc giá mở cửa,


```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    # Giả sử giá mở cửa là 10k, mỗi km là 12k, mỗi phút chờ là 1k
    gia_mo_cua = 10000
    gia_km = 12000
    gia_cho = 1000 
    tong_tien = gia_mo_cua + (km * gia_km) + (wait_minutes * gia_cho) 
    # Nếu đi vào giờ cao điểm (ví dụ 17h-19h) thì tăng thêm tiền chẳng hạn
    if 17 <= start_hour <= 19:
        tong_tien = tong_tien * 1.2 # Tăng 20%   
    return tong_tien
```
