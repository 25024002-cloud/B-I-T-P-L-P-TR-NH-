Nhập vào 3 số nguyên dương a, b, c. Kiểm tra xem a, b, c có cấu thành độ dài cạnh của 1 tam giác được không.
Nếu a, b, c cấu tạo thành được một tam giác, in ra "Yes". Ngược lại in ra "No". 


```python
a, b, c = map(int, input().split())
if a+b>c and a+c>b and b+c>a:
    print("Yes")
else:
    print("No")
```

Kiểm tra xem một số có phải là lũy thừa của 2 hay không
Sử dụng các toán tử bitwise để xác định xem một số có phải là lũy thừa của 2
hay không.
Gợi ý: Đối với lũy thừa của 2, n & (n-1) == 0


```python
n = int(input())
if n > 0 and (n & (n - 1)) == 0:
    print("True")
else:
    print("False")
```

Nhập vào 3 số nguyên dương a,b,c là độ dài 3 cạnh của một tam giác. Kiểm tra 3 số đã nhập có tạo thành một tam giác hợp
lệ hay không. Nếu không in ra "Không phải tam giác". Nếu có, phân loại tam giác đó là "Tam giác đều", "Tam giác cân", hay
"Tam giác thường

```python
a, b, c = map(int,input().split())
if a + b >c and a+c>b and c+b >a:
    if a == b == c:
        print("Tam giac deu")
    elif a==b or a==c or b==c:
        print("Tam giac can")
    else:
        print("Tam giac thuong")
else:
    print("Khong phai tam giac")
```

Nhập vào một số nguyên dương n là số năm, kiểm tra xem năm đó có phải là năm nhuận hay không. Năm nhuận là năm
chia hết cho 4 nhưng không chia hết cho 100, hoặc chia hết cho 400. Nếu là năm nhuận in ra "Yes", nếu không in ra "No" 


```python
n = int(input())
if n % 4 == 0 and n % 100 != 0 or n % 400 == 0:
    print("Yes")
else:
    print("No")
```

Nhập vào một số thực: in ra kết quả làm tròn lên ,tròn xuống, làm tròn tới số nguyên gần nhất (không được sử dụng hàm
sẵn có của pyton). Các kết quả cách nhau một dấu cách 


```python
x = float(input())
# Tròn lên: Nếu là số dương mà có phần lẻ thì cộng 1 rồi cắt đuôi
if x > int(x):
    t_len = int(x) + 1
else:
    t_len = int(x)
# Tròn xuống: Nếu là số âm mà có phần lẻ thì trừ 1 rồi cắt đuôi
if x < int(x):
    t_xuong = int(x) - 1
else:
    t_xuong = int(x)
# Gần nhất: Mẹo đẩy 0.5 
if x >= 0:
    g_nhat = int(x + 0.5)
else:
    g_nhat = int(x - 0.5)

print(f"tron len = {t_len} ,tron xuong = {t_xuong},gan nhat = {g_nhat}")
```

Nhập vào điểm trung bình của một học sinh và in ra học lực của học sinh đó.
Xếp loại học lực theo quy tắc:
-  >= 8.0: Giỏi
-  >= 6.5: Khá
-  >= 5.0: Trung bình
-  < 5.0: Yếu

```python
n = float(input())
if n >= 8:
    print("Giỏi")
elif n >= 6.5:
    print("Khá")
elif n >= 5:
    print("Trung bình")
else:
    print("Yếu")
```
