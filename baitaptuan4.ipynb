```python
while True:
    dong = input()
    if dong == "":
        break
    a, b = map(int, dong.split())
    tong_snt = 0
    for so in range (a, b+1):
        if so >= 2:
            la_snt = True
            for i in range(2, so):
                if so % i == 0:
                    la_snt = False
                    break
            if la_snt == True:
                    tong_snt = tong_snt + so
    print (tong_snt)
```

W4A9 Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số
đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu
cách). Nếu không có, in ra (no number).

W4A10 Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất
và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số
nhỏ nhất."

```python
n = int(input())
max_buoc = 0
so_tot_nhat = 0

for i in range(1, n + 1):
    hien_tai = i
    buoc = 1
    
    # Tính độ dài dãy Collatz của số i
    while hien_tai != 1:
        if hien_tai % 2 == 0:
            hien_tai = hien_tai // 2
        else:
            hien_tai = 3 * hien_tai + 1
        buoc = buoc + 1
    
    # Nếu tìm được dãy dài hơn thì cập nhật kỷ lục
    # (Dùng dấu > đảm bảo nếu có 2 số cùng độ dài, nó sẽ giữ lại số nhỏ hơn tìm được lúc đầu)
    if buoc > max_buoc:
        max_buoc = buoc
        so_tot_nhat = i

print(so_tot_nhat, max_buoc)
```

Nếu có, in ra màn hình "true", ngược lại, in ra "false"

```python
a, b = map(int, input().split())
if a<=0 or b<=0:
    print("false")
else:
    tong_uoc_a = 0
    for i in range(1, a):
        if a % i ==0:
            tong_uoc_a = tong_uoc_a + i
    tong_uoc_b = 0
    for j in range(1, b):
        if b % j== 0:
            tong_uoc_b = tong_uoc_b + j
    if tong_uoc_b == a and tong_uoc_a == b:
        print("true")
    else:
        print("false")

```
