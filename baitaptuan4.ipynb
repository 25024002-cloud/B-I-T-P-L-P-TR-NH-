Viết chương trình Python sử dụng một trong 2 loại vòng lặp for hoặc while để
tính tống các số nguyên từ 1 đến n. (n <= 1000) 


```python
n = int(input())
if n <= 1000:
    tong = 0
    for i in range(1, n+1):
        tong = tong + i
    print (tong)
else:
    print()
```

```python
n = int(input())

if n <= 1000:
    tong = 0
    i = 1 
    while i <= n:
        tong = tong + i
        i = i + 1 
    print("Tổng là:", tong)
```
