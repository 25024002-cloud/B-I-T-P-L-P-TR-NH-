Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là độ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn hình.


```python
import math
a, b, c = map(int, input().split())
if a + b > c and a+c>b and b+c>a:
 if a==b or a==c or b==c:
    print("Tam giac can")
 elif a==b==c:
       print("Tam giac deu")
 elif  a**2+b**2 ==c**2 or a**2+c**2 ==b**2 or c**2+b**2 ==a**2 :
       print("Tam giac vuong")
 else:
       print("Tam giac thuong")
 print(f"Chu vi tam giac la: {a+b+c}")
 p = (a+b+c)/2
 dien_tich= math.sqrt((p-a)*(p-b)*(p-c)*p)
 print(f"Dien tich tam giac la: {dien_tich:.2f}")
else:
 print("Khong la ba canh cua mot tam giac")
```

Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
import math
a, b, c = map(int, input().split())
p= b**2 - 4*a*c
if p < 0: 
    print("Phuong trinh vo nghiem")
elif p == 0:
    print(f"Phuong trinh co nghiem kep : x={-b/(2*a)}")
else:
    print(f"Phuong trinh co nghiem: x1={(-b + math.sqrt(p))/(2*a)}")
    print(f"Phuong trinh co nghiem: x2={(-b - math.sqrt(p))/(2*a)}")
```

Bài 8: Giải hệ phương trình tuyến tính:

ax + by = m
 và cx + dy = n


```python
# Hệ thức Cramer
import math
a, b, m, c,d , n = map(float, input().split())
D = a*d-b*c
Dx = m*d - b*n
Dy = a*n - m*c
if D != 0:
    x = Dx/D
    y = Dy/D
    print(f"He phuong trinh co nghiem duy nhat: x={x}, y={y}")
elif Dx == 0 and Dy == 0:
    print("He phuong trinh vo so nghiem")
else:
    print("He vô nghiem")
```

Bài 9: Nhập vào thời gian 1 công việc nào đó là x giây. Hãy chuyển đổi và viết ra màn hình số thời gian trên dưới dạng bằng bao nhiêu giờ, bao nhiêu phút, bao nhiêu giây. 

```python
# toán tử // (chia lấy nguyên) và % (chia lấy dư).
x = int(input())
gio = x // 3600
giay_con_lai = x % 3600
phut = giay_con_lai // 60
giay = giay_con_lai % 60
print(f"Ket qua: {gio} gio, {phut} phut, {giay} giay ")
```

Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vào một điểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
import math
x, y, R = map(float, input().split())
a,b = map(float, input().split())
if (x-a)**2 + (y-b)**2 != R**2 :
    print("Điem A k thuoc duong tron")
else: 
    print("Điem A thuoc duong tron")
```
